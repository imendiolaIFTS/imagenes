##Ejercicio 1: Espacios de Color

Usando una imagen a color de su elecci√≥n:

a) Cargar la imagen con OpenCV y mostrar los canales BGR por separado. 

```python
#Cargando Imagen
img = cv2.imread('HulkSpiderman.jpg')
#Obteniendo los canales
```

```python
imgG = img[:,:,1] 
imgB = img[:,:,0]
imgR = img[:,:,2]  
plt.imshow(imgG, cmap='gray')
plt.show()
```

```python
def ScaleGray(img: np.ndarray, grises: int) -> np.ndarray:
    d = int(256/grises)
    return (img // d ) *d
```

```python
imgNegStacked = np.hstack([ScaleGray(imgG,2), ScaleGray(imgG,32), ScaleGray(imgG,128)])
plt.imshow(imgNegStacked, cmap='gray')
plt.show()
```

```python
def imgHistogram (img: np.ndarray) -> np.ndarray:
    X = np.zeros(256)
    for i in range(256):
        X[i] = np.count_nonzero(img == i)
    return X
```

```python
x = np.array([[234,12,0],[34,56,16],[233, 112, 9]])
print((x // 2)*2)
```

```python
plt.plot(imgHistogram(imgG), color='green')
plt.plot(imgHistogram(imgB), color='blue')
plt.plot(imgHistogram(imgR), color='red')
plt.title('Histograma')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.grid(True)
plt.show()
```

```python
def undersample(img: np.ndarray, d:int) -> np.ndarray:
    return img[::d,::d]
```

```python
plt.imshow(undersample(imgG,4), cmap='gray')
plt.show()
```
